# histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Load the datasets
train_x = pd.read_csv('X_train_NHkHMNU.csv')
train_y = pd.read_csv('y_train_ZAN5mwg.csv')

# Filter columns with dtype float64
numerical_cols = train_x.select_dtypes(include=['float64']).columns
n_cols = len(numerical_cols)

# Calculate the number of rows and columns for the subplot grid
n_rows = int(np.ceil(np.sqrt(n_cols)))
n_cols = n_rows if n_rows * (n_rows - 1) < n_cols else n_rows - 1

# Create a figure and axes for the subplots
fig, axs = plt.subplots(n_rows, n_cols, figsize=(20, 15))

# Flatten the array of axes (for easy iteration)
axs = axs.flatten()

for i, col in enumerate(numerical_cols):
    sns.histplot(train_x[col], kde=True, ax=axs[i])
    axs[i].axvline(train_x[col].mean(), color='red', linestyle='--', label='Mean')
    axs[i].axvline(train_x[col].median(), color='green', linestyle='-', label='Median')
    axs[i].set_title(f'Distribution of {col}')
    axs[i].set_xlabel(col)
    axs[i].set_ylabel('Frequency')
    axs[i].legend()

# Hide any unused subplots
for ax in axs[len(numerical_cols):]:
    ax.set_visible(False)

# Adjust layout for better visual
plt.tight_layout()

# Save the figure to a file
plt.savefig('numerical_columns_distribution.png')

# Optionally, display the plot
plt.show()
```

```python
# Calculate correlation matrix
corr_matrix = train_x[numerical_cols].corr()

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Heatmap of Correlation Among Numerical Features')
plt.show()
```

```python
# Calculate correlation matrix
corr_matrix = train_x[numerical_cols].corr()

# Identify pairs of columns with high correlation
threshold = 0.6
highly_corr_pairs = []

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            col_pair = (corr_matrix.columns[i], corr_matrix.columns[j], corr_matrix.iloc[i, j])
            highly_corr_pairs.append(col_pair)

# Convert list of tuples into a DataFrame for easier visualization/manipulation
highly_corr_pairs_df = pd.DataFrame(highly_corr_pairs, columns=['Column 1', 'Column 2', 'Correlation'])
print(highly_corr_pairs_df)
```

```python
# Filter the correlation matrix
filtered_corr_matrix = corr_matrix.loc[highly_corr_pairs_df['Column 1'].unique(), highly_corr_pairs_df['Column 2'].unique()]

# Since the filtered matrix might not be square, we make it symmetric for visualization
symmetric_corr_matrix = filtered_corr_matrix.copy()
for col1, col2, _ in highly_corr_pairs:
    symmetric_corr_matrix.loc[col2, col1] = symmetric_corr_matrix.loc[col1, col2]

# Plot heatmap of highly correlated columns
plt.figure(figsize=(12, 8))
sns.heatmap(symmetric_corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Heatmap of Highly Correlated Numerical Features')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

selected_pairs = [('FR_WIND', 'FR_CONSUMPTION'), ('FR_RAIN', 'FR_CONSUMPTION')]

#selected_pairs = highly_corr_pairs_df[['Column 1', 'Column 2']].values[:2]

for col1, col2 in selected_pairs:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=train_x[col1], y=train_x[col2])
    plt.title(f'Relationship between {col1} and {col2}')
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.show()
```

```python
# Replace 'train_x' and 'col1', 'col2' with actual DataFrame and column names
x_clean = train_x[col1].replace([np.inf, -np.inf], np.nan).dropna()
y_clean = train_x[col2].replace([np.inf, -np.inf], np.nan).dropna()

common_indices = x_clean.index.intersection(y_clean.index)
x_clean = x_clean.loc[common_indices]
y_clean = y_clean.loc[common_indices]
```

```python
coeffs = np.polyfit(x_clean, y_clean, 2)
poly_eqn = np.poly1d(coeffs)
x_poly = np.linspace(x_clean.min(), x_clean.max(), 100)
y_poly = poly_eqn(x_poly)

# Scatter plot with polynomial regression line
plt.figure(figsize=(10, 6))
plt.scatter(x_clean, y_clean, label='Data Points')
plt.plot(x_poly, y_poly, color='red', label='Quadratic Fit')
plt.title(f'Quadratic Relationship between {col1} and {col2}')
plt.xlabel(col1)
plt.ylabel(col2)
plt.legend()
plt.show()
```

# divide FR and DE

```python
df_fr = train_x[train_x['COUNTRY']=='FR']
df_de = train_x[train_x['COUNTRY']=='DE']
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Setting 'DAY_ID' as index
'''df_fr.set_index('DAY_ID', inplace=True)
df_de.set_index('DAY_ID', inplace=True)
'''
# Perform seasonal decompose
result_fr = seasonal_decompose(df_fr['FR_CONSUMPTION'], model='additive', period=30)
result_de = seasonal_decompose(df_de['DE_CONSUMPTION'], model='additive', period=30)

# Get common y-limits for the plots
trend_ylim = (min(result_fr.trend.min(), result_de.trend.min()), 
              max(result_fr.trend.max(), result_de.trend.max()))
seasonal_ylim = (min(result_fr.seasonal.min(), result_de.seasonal.min()), 
                 max(result_fr.seasonal.max(), result_de.seasonal.max()))
resid_ylim = (min(result_fr.resid.min(), result_de.resid.min()), 
              max(result_fr.resid.max(), result_de.resid.max()))

# Plotting FR consumption
plt.figure(figsize=(14, 6))
plt.subplot(411)
plt.plot(df_fr['FR_CONSUMPTION'], label='Original')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(result_fr.trend, label='Trend')
plt.ylim(trend_ylim)
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(result_fr.seasonal, label='Seasonal')
plt.ylim(seasonal_ylim)
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(result_fr.resid, label='Residual')
plt.ylim(resid_ylim)
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()

# Plotting DE consumption
plt.figure(figsize=(14, 6))
plt.subplot(411)
plt.plot(df_de['DE_CONSUMPTION'], label='Original')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(result_de.trend, label='Trend')
plt.ylim(trend_ylim)
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(result_de.seasonal, label='Seasonal')
plt.ylim(seasonal_ylim)
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(result_de.resid, label='Residual')
plt.ylim(resid_ylim)
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
